CLASS SSD WHICH ENCAPSULATES ALL SSD FUNCTIONS


```
import subprocess
import os

class SSD:
  def __init__(self):
    subprocess.Popen(["cd", "/content/ssd-custom-dataset"] ,shell=True, 
                                  stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    
    return 
  def train(self,dataset="COCO",batch_size = 2, lr = 1e-5, resume = "weights/vgg16_reducedfc.pth", start_iter = 0):
    #python -W ignore train.py --dataset COCO --batch_size 2 --lr 1e-5 --resume "weights/ssd300_COCO_10000.pth" --start_iter 10000
    
    cmd_list = ['python', '-W', 'ignore', 'train.py', '--dataset', dataset,'--batch_size',str(batch_size),'--lr',str(lr),'--resume', resume,'--start_iter', str(start_iter)]
    result = subprocess.run(cmd_list, stdout=subprocess.PIPE, stderr = subprocess.PIPE)
    print("Training")
    print(result.stderr.decode())
    print (result.stdout.decode())
    return 
  def test(self,model = 'weights/vgg16_reducedfc.pth'):
    #python test.py --trained_model "weights/COCO (1).pth"
    result = subprocess.run(['python','test.py','--trained_model',model], stdout=subprocess.PIPE, stderr = subprocess.PIPE)
    print("Testing")
    print( result.stderr.decode())
    print (result.stdout.decode())
    result = subprocess.run(['python','coco_eval.py'], stdout=subprocess.PIPE, stderr = subprocess.PIPE)
    print("COCO Evaluation")
    print( result.stderr.decode())
    print (result.stdout.decode())
    return
  def test_one(self,model='weights/vgg16_reducedfc.pth',image="11003-Ravi-Singh-Baroda-Gujarat-1_jpg.rf.bd00f49465c1537175003f7c866c64ac.jpg"):
    #!python test_one.py --trained_model "weights/COCO (1).pth" --image "11003-Ravi-Singh-Baroda-Gujarat-1_jpg.rf.bd00f49465c1537175003f7c866c64ac.jpg"
    cmd_list = ['python', 'test_one.py', '--trained_model', model, '--image', image]
    result = subprocess.run(cmd_list, stdout=subprocess.PIPE, stderr = subprocess.PIPE)
    print("Testing sample image")
    print(result.stderr.decode())
    print (result.stdout.decode())
    return
```
